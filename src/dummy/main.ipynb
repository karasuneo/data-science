```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('../../data/1.03. Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x)
results = sm.OLS(y,x).fit()
results.summary()

# F-statistic(F値)からこの分析は有意であるとわかる
# P値をみてもダミー変数も有意であることがわかる  
#　Adj. R-squared(自由度調整済み決定係数)は0.406であることから、このモデルは40.6%の分散を説明できることがわかる 
```

## 出席率が75％以上の人の回帰直線、出席率が75％未満の人の回帰直線をそれぞれ求める

```python
plt.scatter(data['SAT'], y, c=data['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014 * data['SAT']
yhat_yes = 0.8665 + 0.0014 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50026')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

## 出席率を加味しない回帰直線を引く

```python
plt.scatter(data['SAT'], y, c=data['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014 * data['SAT']
yhat_yes = 0.8665 + 0.0014 * data['SAT']
yhat = 0.2758 + 0.0017 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837', label='regression line1')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50026', label='regression line2')
fig = plt.plot(data['SAT'], yhat, lw=3, c='#4C72B0', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

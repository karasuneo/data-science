```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

sns.set()

from scipy import stats

stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

```python
raw_data = pd.read_csv("../../../data/2.02. Binary predictors.csv")
raw_data
```

```python
data = raw_data.copy()
data["Admitted"] = data["Admitted"].map({"Yes": 1, "No": 0})
data["Gender"] = data["Gender"].map({"Female": 1, "Male": 0})
data
```

従属変数(合否)と独立変数(性別)を定義

```python
y = data["Admitted"]
x1 = data[["SAT", "Gender"]]
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y, x)
results_log = reg_log.fit()
results_log.summary()

# Null対数尤度とP値をみると、モデルは有意であることがわかる
# log(odds2) = -0.64 + 2.08 * Gender2 -2
# log(odds1) = -0.64 + 2.08 * Gender1 -1
# 1-2より、log(odds2/odds1) = 2.08(Gender2 - Gender1)
# Gender2 = 1, Gender1 = 0とすると、log(odds2/odds1) = 2.08
# 対数の定理より oddfemale = 7.99 * oddmale
# つまり、性別は大学の合否に影響していることがわかる
```

```python
np.exp(1.9449)
# 具体的には同じSATスコアの場合、女性の方が8倍合格しやすい
```

予測された値 (0.~が混じっているのは確率を示しているため)

```python
np.set_printoptions(formatter={"float": lambda x: "{0:0.2f}".format(x)})
results_log.predict()
```

混同行列

```python
cm_df = pd.DataFrame(results_log.pred_table())
cm_df.columns = ["Predicted 0", "Predicted 1"]
cm_df = cm_df.rename(index={0: "Actual 0", 1: "Actual 1"})
cm_df

# 69は作成したモデルが0、0を予測した数
# 5は1と予測した値が実際に0であった数
```

```python
cm = np.array(cm_df)
accuracy_train = (cm[0, 0] + cm[1, 1]) / cm.sum()
accuracy_train
```

#### モデルにまだ使われていないデータを使って現状作ったモデルをテストしてみる

```python
test = pd.read_csv("../../../data/2.03. Test dataset.csv")
test
```

```python
test["Admitted"] = test["Admitted"].map({"Yes": 1, "No": 0})
test["Gender"] = test["Gender"].map({"Female": 1, "Male": 0})
test
```

```python
test_actual = test["Admitted"]
test_data = test.drop(["Admitted"], axis=1)
test_data = sm.add_constant(test_data)

test_data
```

```python
def confusion_matrix(data, actual_values, model):
    pred_values = model.predict(data)
    bins = np.array([0, 0.5, 1])
    cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
    accuracy = (cm[0, 0] + cm[1, 1]) / cm.sum()
    return cm, accuracy
```

混合行列を作成
- 一般的に訓練用のデータの方がテスト用のデータよりも精度が高い
  - 過学習によるもの

```python
cm = confusion_matrix(test_data, test_actual, results_log)
cm
```

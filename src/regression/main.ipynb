```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()


data = pd.read_csv('../../data/1.01. Simple linear regression.csv')

data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']


plt.scatter(x1,y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python

# stasmodelに変数を追加する
x = sm.add_constant(x1)
# 最小二乗回帰式
result = sm.OLS(y,x).fit()
# 結果を表示
result.summary()

# y^ = b1x1 + b0
# y^ = 0.0017x1 + 0.275
# 二つ目のグラフのcoef(=b0)は切片

# 二つ目のグラフのP値が0.05より小さいので有意 
```

```python
# 観測された点に最も近い回帰直線を引く
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
# yhat = 0*x1 + 0.275
fig = plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.xlim(0)
plt.ylim(0)
plt.show()
```
